```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('james_bond_data_cleansed.csv')
print(data)
```

```python
print(data.info())
```

```python
print(data['imdb'])
```

```python
plt.scatter(data['imdb'], data['rotten_tomatoes'])
plt.xlabel('Average IMDb Rating')
plt.ylabel('Average Rotten Tomatoes Rating')
plt.show()
```

```python
# There is a correlation because for most data points, the more you go to the right,
# the higher up in the graph the data points go, meaning as one increases for one
# condition, the other also increases
```

```python
plt.scatter(data['imdb'], data['bond_kills'])
plt.xlabel('IMDb Rating')
plt.ylabel('Bond Kills')
plt.show()
```

```python
# I would say there isn't a correlation between IMDb Rating and the amount of bad
# guys Bond kills because the data is more scattered around than the last graph
# and doesn't show a pattern ot trend it follows.
```

```python
plt.plot(data['release_year'], data['movie_budget'], label='Movie Budget')
plt.plot(data['release_year'], data['income_world'], label='Income World')
plt.legend()
plt.xlabel('Release Year')
plt.show()
```

```python
# The trends I notice from this graph are that the income world was increasing at a
# significant rate and reached a peak near 2010 and has been declining since. The trend
# I notice for the movie budget is that it has stayed relatively similar from 1960 to
# 2020 and hasn't seen such a drastic increase like the income world. We can also note
# from this graph that the income world is way bigger than the movie budget.
```

```python
income_budget_ratio = data['income_world'] / data['movie_budget']
plt.bar(data['release_year'], income_budget_ratio)
plt.xlabel('Release Year')
plt.ylabel('Income World-Budget Ratio')
plt.show()
```

```python
# The trend I notice here is that the producers had a massive world income-budget ratio
# and overtime this number has drastically fallen. This may be due to more budget being
# spent on the movies or less world income being generated. From this trend, I do not
# thing the producers make more profit than they did in the past because the ratio is
# much smaller now.
```

```python
import statistics
print(min(data['film_length']))
print(max(data['film_length']))
fl_mean = sum(data['film_length']) / len(data['film_length'])
print(fl_mean)
print(statistics.median(data['film_length']))
```

```python
# The shortest film length is 106 minutes. James Bond Quantum of Solace.
# The longest film length is 163 minutes. James Bond No Time to Die
# The mean or average film length is 128.28 minutes.
# The median film length is 128 minutes. James Bond The World Is Not Enough
```
